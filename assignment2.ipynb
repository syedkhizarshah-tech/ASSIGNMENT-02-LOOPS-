# 4. Write a program that takes a number n as input from the user and generates the first n terms of the series formed by squaring the natural numbers. Sample output Enter a number: 6 The first 6 terms of the series are: 1 4 9 16 25 36. 

```python
n = int(input("Enter a number: "))
squares = []
for i in range(1, n + 1):
    squares.append(i**2)
print(f"The first {n} terms of the series are: {' '.join(map(str, squares))}")
```

# 5. Write a program that prompts the user to input a number and prints its multiplication table.

```python

n = int(input("Enter a number: "))
range_limit = int(input("Enter the range limit (default is 10): "))
print(f"\nMultiplication table of {n} (up to {range_limit}):")
for i in range(1, range_limit + 1):
    print(f"{n} × {i} = {n * i}")
```

# 6. Write a program that prompts the user to enter a number n, and then prints all the odd numbers between 1 and n.

```python
n = int(input("Enter a number: "))
odd_numbers = [i for i in range(1, n + 1) if i % 2 != 0]
print(f"Odd numbers between 1 and {n}:")
print(" ".join(map(str, odd_numbers)))
```

# 7. Write a Python program to print the numbers from 20 to 1 using a while loop.

```python
start = int(input("Enter starting number: "))
print(f"Numbers from {start} to 1:")
while start >= 1:
    print(start, end=" ")
    start -= 1
```

# 9. Write a program that prompts the user to enter a number and repeats this process 5 times. The program should accumulate the numbers entered and then display the final running total.
# Sample Output:
# Enter a number: 10
# Enter a number: 15
# Enter a number: 35
 # Enter a number: 40
# Enter a number: 50
# The final running total is: 150.


```python
# Program to accumulate numbers entered by the user

# Initialize the running total
total = 0

print("You will be asked to enter 5 numbers.")
print("Each number will be added to the running total.\n")

# Repeat the process 5 times
for i in range(1, 6):  # Loop from 1 to 5
    number = int(input(f"Enter number {i}: "))
    total += number  # Add the entered number to the total
    print(f"After entry {i}, the running total is: {total}\n")

# Display the final running total
print("===================================")
print("All numbers have been entered.")
print("The final running total is:", total)
print("===================================")
```

# 10. Write a Python program that:
## 1- Creates a list named clean_countries containing the following countries: [Bangladesh India Pakistan China]

## 2- Repeatedly asks the user to enter a country name.

## 3- Checks whether the entered country exists in the clean_countries list.

## 4-If the country is found in the list:

## Display the message:

## <country_name> is in clean countries list

## Stop the program.

## 5-If the country is not found, keep asking the user to enter another country.



```python
# clean_countries lookup program
# The program repeatedly asks the user for a country name
# and stops when a country from the clean_countries list is entered.

clean_countries = ["Bangladesh", "India", "Pakistan", "China"]

print("This program will keep asking you to enter a country name.")
print("It will stop only when you enter a country that is in the clean_countries list.\n")

while True:
    # Read input from the user
    user_input = input("Enter a country name: ")

    # Normalize input: remove extra spaces and make case consistent
    country = user_input.strip()

    # Check for empty input and prompt again if needed
    if country == "":
        print("You entered nothing. Please type a country name.\n")
        continue

    # Check membership in a case-insensitive way
    found = any(country.lower() == c.lower() for c in clean_countries)

    if found:
        print(f"{country} is in clean countries list")
        break
    else:
        print(f"{country} is not in the list. Please try again.\n")
```
